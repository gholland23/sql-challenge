```python
import sqlalchemy
import pandas as pd
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
username = 'postgres'
password = 'postgres'
dbname = 'HOMEWORK'
engine = create_engine(f'postgres://{username}:{password}@localhost/{dbname}')
```

```python
connect = engine.connect()
```

```python
inspector = inspect(engine)
```

```python
inspector.get_table_names()
```

```python
columns = inspector.get_columns('salaries')
for column in columns:
    print(column['name'], column['type'])
```

```python
sal_df = pd.read_sql_table('salaries',connect)
sal_df
```

```python
salary_df = pd.DataFrame(sal_df)
salary_df.drop_duplicates("emp_no")
```

```python
columns = inspector.get_columns('titles')
for column in columns:
    print(column['name'], column['type'])
```

```python
titles_df = pd.read_sql_table('titles',connect)
titles_df
```

```python
Titles_df = pd.DataFrame(titles_df)
Titles_df.drop_duplicates("emp_no")
```

```python
merged_df = pd.merge(salary_df, Titles_df, how = 'left', on='emp_no')
Merged_df = merged_df.drop_duplicates("emp_no")
Merged_df.set_index("emp_no")
```

```python
#plot
grouped_data = Merged_df.groupby('title').mean()['salary']
grouped_data_rounded = round(grouped_data,2)
grouped_data_rounded
```

```python
ST_df = pd.DataFrame(grouped_data_rounded)
ST_df
```

```python
plot = ST_df.plot.bar(rot=70)
```

```python
#salaries to list for histogram
salaries = merged_df['salary'].tolist()
```

```python
#histogram
x = salaries
num_bins = 5
plto = plt.hist(x, num_bins, facecolor= "blue", alpha = 0.5)
plt.ylabel("salary")
```
